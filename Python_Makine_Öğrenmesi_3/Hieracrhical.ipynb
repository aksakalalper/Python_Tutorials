```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram,linkage
x,_=make_blobs(n_samples=300,random_state=42,centers=4,cluster_std=0.5)
```

```python
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python

linkage_methods = ['single', 'average', 'complete', 'ward']
fig, axes = plt.subplots(2, 4, figsize=(15, 8))  # 2x4 dzeninde subplot oluturma

for i, linkage_method in enumerate(linkage_methods):
    # Metodlar retilir
    ac = AgglomerativeClustering(n_clusters=4, linkage=linkage_method)
    cluster_labels = ac.fit_predict(x)
    
    # Dendrogram iin uygun veriyi hesaplama
    linked = linkage(x, method=linkage_method)
    
    # Veri noktalar ve uzaklk grafii
    dendrogram(linked, ax=axes[0, i], no_labels=True)
    axes[0, i].set_title(linkage_method)
    
    # x ve y grafii
    axes[1, i].scatter(x[:, 0], x[:, 1], c=cluster_labels, cmap='viridis')
    axes[1, i].set_title(linkage_method)

plt.tight_layout()
plt.show()


```
