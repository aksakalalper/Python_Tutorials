```python
import pandas as pd
import numpy as np  
import os
from sklearn.model_selection import train_test_split   
import tensorflow as tf
import keras
```

Introduction to ML


```python
os.chdir("D:\Python_Tutorials\Deep_Learning\Docs")

data = pd.read_csv('Folds5x2_pp.csv')
df=pd.DataFrame(data)
df.head()
```

```python
y=df['PE']
X=df.drop('PE',axis=1)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
ann=tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=6,activation='relu',input_dim=4))
ann.add(tf.keras.layers.Dense(units=6,activation='relu'))
ann.add(tf.keras.layers.Dense(units=1)) #if we want to classification we should use sigmoid activation function 
ann.compile(optimizer='adam',loss='mean_squared_error')
ann.fit(X_train,y_train,batch_size=32,epochs=100)
```

```python
train_loss = ann.evaluate(X_train, y_train)
print(f'Train Loss: {train_loss}')

test_loss = ann.evaluate(X_test, y_test)
print(f'Test Loss: {test_loss}')
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score

y_pred = ann.predict(X_test)

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"Mean Absolute Error: {mae}")
print(f"R^2 Score: {r2}")
```

```python
from sklearn.preprocessing import StandardScaler

ann.save('ann_model.h5')

def predict(feature1, feature2, feature3, feature4):
    model = tf.keras.models.load_model('ann_model.h5')
    input_data = np.array([[feature1, feature2, feature3, feature4]])
    input_scaled = (input_data)
    prediction = model.predict(input_scaled)
    return prediction[0][0]

feature1 = 23.64							
feature2 = 58.49
feature3 = 1011.40
feature4 = 74.20

predicted_value = predict(feature1, feature2, feature3, feature4)
print(f"Predicted Value: {predicted_value}")
```
