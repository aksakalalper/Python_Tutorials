importing libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

reading file

```python
file="athlete_events_cleaned.csv"
data=pd.read_csv(file)
df=pd.DataFrame(data)
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
plt.plot(df["Age"].head(20),color="red")
plt.xlabel("Line")
plt.ylabel("Age")
plt.show()
```

```python
plt.figure(figsize=(20,10))
plt.subplot(131)
plt.hist(df["Height"].head(20),color="red",width=1)
plt.xlabel("Height")
plt.ylabel("Count")
plt.subplot(132)
plt.xlabel("Weight")
plt.ylabel("Count")
plt.hist(df["Weight"].head(20),color="red",width=1)
plt.subplot(133)
plt.xlabel('Height')
plt.ylabel('Weight')
plt.scatter(df['Height'].head(20),df['Weight'].head(20))
plt.tight_layout()
plt.show()
```

Scatter plot

```python
sns.scatterplot(x="Age",y="Height",data=df,hue="Sex",size="Height")
sns.set_style('dark')
plt.show()
```

Regplot

```python
sns.regplot(x="Weight",y="Height",data=df,marker='+',scatter_kws={'alpha':0.1})
plt.show()
```

Line plot

```python
sns.lineplot(x="Height",y="Weight",data=df,hue="Sex")   
plt.show()
```

hist plot

```python
import seaborn as sns
```

```python
sns.histplot(df["Height"])
plt.show()
```

```python
sns.displot(df,x="Height",y="Weight",hue="Sex",kind="kde")
plt.show()
```

bar plot

```python
sns.barplot(x="Medal",y="Weight",data=df)
plt.show()
```

box plot

```python
sns.boxplot(x="Medal",y="Height",data=df)
plt.show()
```

violin plot

```python
sns.violinplot(x="Medal",y="Height",hue="Sex",data=df)
plt.show()
```

heat map


```python
sns.heatmap(df[['Height','Weight']].corr(),cmap="YlGnBu",annot=True)
plt.show()
```

pair plot


```python
sns.pairplot(df)
plt.show()
```

count plot  

```python
plt.figure(figsize=(20,10))
sns.countplot(x="City",data=df)
plt.xticks(rotation=90)
plt.show()
```
