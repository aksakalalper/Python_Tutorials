K-Means Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import  KMeans
#cluster1
x1=np.random.normal(25,5,1000)
y1=np.random.normal(25,5,1000)
#cluster2
x2=np.random.normal(55,5,1000)
y2=np.random.normal(55,5,1000)
#cluster3
x3=np.random.normal(55,5,1000)
y3=np.random.normal(15,5,1000)

x = np.concatenate([x1, x2, x3])
y = np.concatenate([y1, y2, y3])
df = pd.DataFrame({'x': x,'y': y})

wcs=[]

sns.scatterplot(x=x1,y=y1,color='red')
sns.scatterplot(x=x2,y=y2,color='blue')
sns.scatterplot(x=x3,y=y3,color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

Choosing K-elbow

```python
for i in range(1,15):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df)
    wcs.append(kmeans.inertia_)

plt.plot(range(1,15),wcs)
plt.xlabel('cluster count')
plt.ylabel('wcs')
plt.show()
```

Hierarchical Clustering

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.cluster.hierarchy import linkage,dendrogram
from sklearn.cluster import AgglomerativeClustering

#cluster1
x1=np.random.normal(35,5,500)
y1=np.random.normal(25,5,500)
#cluster2
x2=np.random.normal(45,5,500)
y2=np.random.normal(55,5,500)
#cluster3
x3=np.random.normal(55,5,500)
y3=np.random.normal(15,5,500)

x = np.concatenate([x1, x2, x3])
y = np.concatenate([y1, y2, y3])
df = pd.DataFrame({'x': x,'y': y})

wcs=[]

sns.scatterplot(x=x1,y=y1,color='red')
sns.scatterplot(x=x2,y=y2,color='blue')
sns.scatterplot(x=x3,y=y3,color='green')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

Dendrogram 

```python
merg=linkage(df,method='ward')
dendrogram(merg,leaf_rotation=90)
plt.xlabel('data points')
plt.ylabel('oclidian distance')
plt.show()
```

Principal Component Analysis

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA

iris=load_iris()
data=iris.data
feature_names=iris.feature_names
y=iris.target

df=pd.DataFrame(data=data,columns=feature_names)
```

```python
pca=PCA(n_components=2,whiten=True) #whiten means normalize.
pca.fit(df)

x_pca=pca.transform(df)
print('variance ratio ',pca.explained_variance_ratio_)
print('sum: ',pca.explained_variance_ratio_.sum())
```

```python
df['pca1'] = x_pca[:, 0] 
df['pca2'] = x_pca[:, 1]

# Scatter plot ile görselleştirme
plt.figure(figsize=(10, 6))
sns.scatterplot(x='pca1', y='pca2', hue='species', data=df)
plt.title('Iris Veri Seti - PCA İle Görselleştirme')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()


```
